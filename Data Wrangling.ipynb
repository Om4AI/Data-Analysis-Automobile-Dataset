## Handling the missing data

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

<b>Replace by mean:</b>

<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>

<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four". 
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>

<ul>
    <li>"price": 4 missing data, simply delete the whole row
        <ul>
            <li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>
        </ul>
    </li>
</ul>

## Handling the missing data actually

```python
# Convert the normalized losses to float data

autos["normalized-losses"] = autos["normalized-losses"].astype("float")
```

### Replace with mean: ***Normalized Losses***

```python
# Calculate the mean

avg_normloss = autos["normalized-losses"].astype("float").mean()
print("Average normalized losses: ",avg_normloss)

# Replace the mean value in column
autos["normalized-losses"].replace(np.nan, avg_normloss, inplace=True)
```

```python
# autos.dtypes
```

```python
autos.head()
```

### Replace with mean: ***Stroke, Bore, Horse-power, Peak-rpm***

```python
autos["stroke"] = autos["stroke"].astype("float")

avg_stroke = autos["stroke"].astype("float").mean()
print("Average Stroke value: ",avg_stroke)

# Replace
autos["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
autos["bore"] = autos["bore"].astype("float")

avg_bore = autos["bore"].astype("float").mean()
print("Average Bore value: ",avg_bore)

# Replace
autos["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
autos["horsepower"] = autos["horsepower"].astype("float")

avg_horsepower = autos["horsepower"].astype("float").mean()
print("Average Horsepower value: ",avg_horsepower)

# Replace
autos["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
autos["peak-rpm"] = autos["peak-rpm"].astype("float")

avg_peakrpm = autos["peak-rpm"].astype("float").mean()
print("Average Peakrpm value: ",avg_peakrpm)

# Replace
autos["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
```

### Replace with Frequency: ***Num_doors***

```python
autos["num-of-doors"].value_counts()
```

```python
autos["num-of-doors"].replace(np.nan, "four", inplace=True)
```

<h2 id="data_standardization">Data Standardization</h2>
<p>
Data is usually collected from different agencies with different formats.
(Data Standardization is also a term for a particular type of data normalization, where we subtract the mean and divide by the standard deviation)
</p>
    
<b>What is Standardization?</b>
<p>Standardization is the process of transforming data into a common format which allows the researcher to make the meaningful comparison.
</p>

<b>Example</b>

<p>Transform mpg to L/100km:</p>
<p>In our dataset, the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we are developing an application in a country that accept the fuel consumption with L/100km standard</p>
<p>We will need to apply <b>data transformation</b> to transform mpg into L/100km?</p>


<p>The formula for unit conversion is<p>
L/100km = 235 / mpg
<p>We can do many mathematical operations directly in Pandas.</p>


```python
autos.head()
```

```python
# Transform
autos["city-L/100km"] = 235/autos["city-mpg"]

# Display data
autos.head()
```

```python
autos["highway-mpg"] = 235/autos["highway-mpg"]

autos.rename(columns={"highway-mpg":"highway-L/100km"}, inplace=True)
autos.head()
```

### Data Normalization:    ***original_data = original_data/max(original_data)***

```python
# Bring all values in the range:0-1

autos["length"] = autos["length"]/autos["length"].max()
autos["width"] = autos["width"]/autos["width"].max()
autos["height"] = autos["height"]/autos["height"].max()
```

```python
autos[["length", "width","height"]].head()
```

## ***Binning***

We are going to divide the **Horsepower** column into 3 categories using the ***Pandas.cut()*** method

```python
autos["horsepower"] = autos["horsepower"].astype("int", copy=True)
```

```python
# Plot HorsePower

plt.figure(figsize=(10,6))
plt.hist(autos["horsepower"], color="orange")
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Binning")
plt.show()
```

```python
# Create Bins

bins = np.linspace(min(autos["horsepower"]), max(autos["horsepower"]), 4)
bins
```

```python
# Group names

group_names = ["Low","Medium","High"]
```

```python
# This introduces a new column in dataframe
autos["horsepower-binned"] = pd.cut(autos["horsepower"], bins, labels=group_names, include_lowest=True)
autos[["horsepower", "horsepower-binned"]].head(10)
```

```python
# Number of vehicles in each bin

autos["horsepower-binned"].value_counts()
```

```python
# Plot binned data

plt.figure(figsize=(8,6))
plt.hist(autos["horsepower-binned"], color="orange")
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Binned")
plt.show()
```

<h2 id="indicator">Indicator variable (or dummy variable)</h2>
<b>What is an indicator variable?</b>
<p>
    An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning. 
</p>

<b>Why we use indicator variables?</b>

<p>
    So we can use categorical variables for regression analysis in the later modules.
</p>
<b>Example</b>
<p>
    We see the column "fuel-type" has two unique values, "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "fuel-type" into indicator variables.
</p>

<p>
    We will use the panda's method 'get_dummies' to assign numerical values to different categories of fuel type. 
</p>


### ***Fuel-type indicator variable***

```python
autos.columns
```

```python
# Get dummy variables & assign it to a dataframe

dummy_variable_1 = pd.get_dummies(autos["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# Merge the new dummy variables & the existing dataframe
autos = pd.concat([autos, dummy_variable_1], axis = 1)

# Drop original column of fueltype
autos.drop("fuel-type", axis=1, inplace=True)
```

```python
autos.head()
```

### ***Aspiration indicator variable***

```python
# Generate dummies

dummy_variable_2 = pd.get_dummies(autos["aspiration"])

# change column names for clarity
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

# show first 5 instances of data frame "dummy_variable_"
dummy_variable_2.head()
```

```python
# merge the new dataframe to the original datafram
autos = pd.concat([autos, dummy_variable_2], axis=1)

# drop original column "aspiration" from "df"
autos.drop('aspiration', axis = 1, inplace=True)
```
